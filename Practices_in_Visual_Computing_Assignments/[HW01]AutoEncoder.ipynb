  + PIL image는 각 픽셀이 [0, 255] 사이값으로 구성된 H x W x C 크기의 이미지이다.

```python
import torchvision.transforms as transforms
from torchvision.datasets import MNIST

# download path 정의
download_root = './MNIST_DATASET'


# PIL image (H x W x C) -> torch tensor (C x H x W) and
# [0, 255] -> [0.0, 1.0]
mnist_transform = transforms.Compose([
    transforms.ToTensor(),
])

train_dataset = MNIST(download_root,
                      transform=mnist_transform,
                      train=True,
                      download=True)

test_dataset = MNIST(download_root,
                     transform=mnist_transform,
                     train=False,
                     download=False)
```

## 네트워크 모델 설계
* 인코더 모델: `torch.nn.Sequential` 모듈로 모델로 설계
  + InputLayer로 (1,28,28) 영상을 받고, Linear 레이어를 여럿 통과시킨 후, 출력으로 n_dim차원 벡터가 나오도록 함.
* 디코더 모델: `torch.nn.Sequential` 모듈로 모델로 설계
  + InputLayer에서 n_dim차원 벡터를 받고, Linear 레이어를 여럿 통과시킨 후, 출력으로 (1,28,28) 영상이 나오도록 함.
* 오토인코더 모델: 인코더, 디코더를 결합하여 설계

잠재 벡터(latent vector) $z$의 크기 설정
여기서는 n_dim을 우선 2로 설정한다.
* 즉, n_dim=2

```python
n_dim = 4
```

인코더 모델 정의
* (1, 28, 28) 영상을 입력으로 받도록 입력 레이어 정의
* Flatten으로 입력 텐서를 784-vector로 벡터라이즈
* Fully connected layer로 784 > 256 > 128 > 32 > n_dim 로 차원 축소

```python
enc = nn.Sequential(
  nn.Flatten(),    # 784 = 1 x 28 x 28
  nn.Linear(784, 256),
  nn.ReLU(),
  nn.Linear(256, 128),
  nn.ReLU(),
  nn.Linear(128, 32),
  nn.ReLU(),
  nn.Linear(32, n_dim)
)

enc = enc.to(device)

print(enc)
```

디코더 모델 정의
* Fully connected layey로 n_dim > 32 > 128 > 256 > 784로 차원 확대
* 784-vector를 Reshape을 통해 (28, 28)의 텐서로 변환

```python
dec = nn.Sequential(
    nn.Linear(n_dim, 32),
    nn.ReLU(),
    nn.Linear(32, 256),
    nn.ReLU(),
    nn.Linear(256, 784),
    nn.Unflatten(1, torch.Size([1, 28, 28])),
    nn.Sigmoid(),
)

dec = dec.to(device)

print(dec)
```

AutoEncoder 모델 정의
* (1, 28, 28) 영상을 입력으로 받도록 입력 레이어 정의
* 입력 레이어 > 인코더 > 디코더로 구성
* (1, 28, 28) 영상이 출력되로록 아웃풋 레이어 정의

```python
ae = nn.Sequential(
    enc,
    dec,
)

ae = ae.to(device)

print(ae)
```

## 2. 결과에 대한 분석을 작성한다.

전체적으로 0.026x ~ 0.028x의 Loss로 비슷한 학습 경향성을 보여준다. 가중치 분산 그래프를 봐도 전체적으로 비슷하지만 아래의 차이가 있다.

- xavier초기화는 He초기화보다 초기화하는 값의 분산이 작으므로, 학습 시 초기화방식을 2, 3번씩 돌렸을 때, 전부 0.27x의 Loss가 나오는 등 학습 안정성이 높다는 특징이 있다.

- He초기화는 xavier초기화보다 초기화하는 값의 분산이 크므로, Loss를 통한 학습 안정성이 xavier보다는 낮다고 볼 수 있다. 실제로 0.026x ~ 0.028x의 결과로 학습 안정성이 xavier보다는 낮다.

- 또한 Normal과 Uniform의 차이는 분포를 추출하는 확률분포의 차이라고 볼 수 있다.

  Normal은 가우시안 분포에서의 추출, Uniform은 특정 구간 내의 균등한 확률분포에서 가중치를 추출한다는 차이가 있다.

  Normal의 경우 정규분포에서 추출한 값을 기준으로 초기화를 진행해 정규분포 모양을 따른다고 볼 수 있다.
  
  Uniform의 경우 따로 설정해주지 않는 경우 일정한 같은 분포에서 균등한 개수의 가중치를 추출한다.

  실험 횟수가 적었기 때문에 학습 시의 유의미한 차이는 구하지 못했지만, 0을 중심으로 0주변에 값이 분포되는 Normal이 학습 안정성이 더 좋을 것이다.

-- 이 때 학습 안정성이 좋다는 뜻은 학습 횟수가 늘어나도 Loss가 비슷한 값으로 수렴한다는 뜻이고, 학습 안정성이 나쁘다는 뜻은 학습 시에 Loss의 분산이 커진다는 이야기(학습 결과가 더 좋을 수도 있고 나쁠 수도 있다)이다.

## 훈련 전, 네트워크 모델을 함수로서 활용
* AutoEncoder ae를 모델로 구성했기 때문에, 지금부터 함수로서 활용 가능
  + 단, ae 함수는 batch 단위로 수행됨을 명심할 것.  
    - 단순히, (1, 28, 28) -> ae -> (1, 28, 28)로 동작하지 않고,
    - batch 단위로 (?, 1, 28, 28) -> ae -> (?, 1, 28, 28)로 병렬처리됨.

## 딥러닝 모델 학습

pytorch에서는 tensorflow기반 코드와는 달리 `fit()` 함수를 직접 구현해야한다.

```python
def fit(data_loader, epochs):
  for epoch in range(epochs):
    now = time.time()
    avg_loss = 0
    total_batch = len(data_loader)

    for i, (batch_images, batch_labels) in enumerate(data_loader):

      X = batch_images.to(device)  # [-1, 1, 28, 28]

      # forward 단계
      # 1. input data를 모델에 통과시킵니다.
      # 2. loss를 계산합니다.
      Y_prediction = ae(X)
      loss = criterion(Y_prediction, X)

      # bacward 단계
      # 1. backprop 단계를 실행하기 전에 변화도를 0으로 만듭니다.
      # 2. 모델의 매개변수에 대한 손실의 변화도를 계산합니다.
      # 3. step 함수를 호출하면 매개변수가 갱신됩니다.
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()

      avg_loss += loss / total_batch

    print("[Epoch: {:>4}] \t loss = {:.4f} \t time = {:.4f}"
          .format(epoch + 1, avg_loss.data, time.time()-now))

  print("Learning Finished!")
```
